```python
df["cabin"].head(20)
df=df.drop(columns=["cabin"])
```

```python
df.fillna(df["age"].mean(),inplace=True)
import numpy as np

df["embarked"] = df["embarked"].apply(
    lambda x: np.nan if isinstance(x, (int, float)) else x
)


df["embarked"].fillna(df["embarked"].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
'''
ðŸŽ¯ PHASE 5: Data Type Fixing

Q15. Convert survived column to category
Q16. Convert sex column to category
Q17. Convert pclass column to category

ðŸ‘‰ Goal: Memory optimization + clarity
'''
```

```python
df["survived"]=df["survived"].apply(lambda x: "survived" if x==1 else "not survived")
df["survived"]=df["survived"].astype("category")
df["sex"]=df["sex"].astype("category")
df["pclass"]=df["pclass"].astype("category")
```

```python
average_fare=df["fare"].mean()
print("average fare",average_fare)
df[df["fare"]>average_fare].head()
```

```python
'''
âž• PHASE 7: Feature Engineering (VERY IMPORTANT)

Q22. Create a column family_size = sibsp + parch + 1
Q23. Create a column is_alone

1 if family_size == 1

0 otherwise

Q24. Create a column fare_per_person
'''
```

```python
df["family_size"]=df["sibsp"]+df["parch"]+1
df["family_size"].head(10)
```

```python
df["is_alone"]=df["family_size"].apply(lambda x: 1 if x==1 else 0)
df["is_alone"].head()
df["is_alone"].astype("category")
```

```python
df["fare_average"]=df["fare"]/df["family_size"]
df["fare_average"].head().round(2)
```

```python
'''
ðŸ“Š PHASE 8: GroupBy Analysis (REAL ANALYSIS)

Q25. Find survival rate by gender
Q26. Find survival rate by passenger class
Q27. Find average age of survivors vs non-survivors
Q28. Find average fare by embarkation port
'''
```

```python
df["survived"].nunique()
```

```python
df["survived"] = df["survived"].replace({
    "not survived": 0,
    "survived": 1
})

df["survived"] = df["survived"].astype(int)
df.groupby("sex")["survived"].mean()
```

```python
df.groupby("pclass")["survived"].mean().round(2)
```

```python
df.groupby("survived")["age"].mean().round(2)
```

```python
df.groupby("embarked")["fare"].mean().round()
```
